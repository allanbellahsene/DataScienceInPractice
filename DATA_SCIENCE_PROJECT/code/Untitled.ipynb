```python
def kelly_criterion(decimal_odd, value_diff):
    true_prob = 1/(decimal_odd - value_diff)
    q = 1 - true_prob
    b = decimal_odd - 1
    optimal_frac = ( b*true_prob - q ) / b
    return optimal_frac
    
```

```python
def mise_fixe_strategy(data, initial_capital, mise):
    profits=[]
    total_capital=[]
    total_capital.append(initial_capital)
    for i in range(1,len(data)):
        odd = data['1.73'][i]
        if data['1'][i] == 1:
            profit = mise * odd
        else:
            profit = -mise
        profits.append(profit)
        capital = total_capital[i-1] + profit
        total_capital.append(capital)
```

```python
def optimal_strategy(data,initial_capital, min_bankroll, bet_advantage, mise=None):
        
    profits=[]
    total_capital = []
    total_capital.append(initial_capital)
    for i in range(1, len(data)):
        odd = data['1.73'][i]
        if not mise:
            mise = kelly_criterion(odd, bet_advantage) * initial_capital
        else:
            mise = mise
            
        if data['-1'][i] == 1:
            profit = mise * (odd-1)
        else:
            profit = -mise
        profits.append(profit)
        capital = total_capital[i-1] + profit
        total_capital.append(capital) 
        if capital < min_bankroll:
            break
                
    total_profit = np.sum(profits)
    strategy_return = total_profit/initial_capital
           
    return strategy_return
```

```python
[total_capital,profits, total_profit, strategy_return] = optimal_strategy(data=data, initial_capital=1000, max_loss=200, bet_advantage=0.0001)
```

```python
[total_capital, profits, total_profit, strategy_return]=optimal_strategy(data=data, initial_capital=1000, max_loss=500, bet_advantage=0.001, mise=50)
```
